```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
'''After importing the required libraries, make sure that your are able to load the image using opencv. OpenCV loads the image as a numpy array. The color of each pixel would by default be sorted as BGR instead of the standard RGB. We can convert the grayscale image back to the normal colored image using cvtColor conversion function provided by openCV. I would be doing so after face detection. You can do it before as well.'''
```

```python
img = cv2.imread('/home/prakarsh/Desktop/group.jpg')
plt.imshow(img)
plt.show()
```

```python
'''The openCV library provides a Haar Cascade (open source) which we would be using. More details about that can be found at http://docs.opencv.org/trunk/d7/d8b/tutorial_py_face_detection.html'''
```

```python
faceCascade = cv2.CascadeClassifier('/home/prakarsh/anaconda2/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray, 1.3, 5)
print faces

for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0), 5)
    roiGray = gray[y:y+h, x:x+w]
    roiColor = img[y:y+h, x:x+w]
plt.imshow(img)
plt.show()
```
